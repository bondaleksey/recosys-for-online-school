# Характеристики пользователя
- способность обучаться (оценка вступительного, посещаемость занятий, выполнение ДЗ, активное участие на занятиях)
- степень удовлетворенности/вовлеченности в прошедшие курсы (лекции, комментарии, отзывы)
- основное "направление"/тематика курсов: Web, DevOps, Backend, Math, Algos ... 
- степень/уровень платежеспособности (проходил бесплатные курсы, покупал дешевые курсы, купил годовую подписку на обучение)

# Характеристика курсов
- рейтинги (популярность/оценки от пользователей, обязательность/востребованность/популярность на рынке, персонализированный рейтинг vs не персонализированный)
- сложность (количество prerequisites, сколько поступило, сколько не закончило)
- специфичность курса (java vs kafka)
- lift на курсы (например, алгоритмы + python/c++/Java) (связка курсов от пользователей по купленным курсам)
- граф связности курсов по программам/prerequisites (от преподавателей по программам)

# Особенности задачи 
- количество пользователей с более чем 1/2 оконченными курсами
- число разных курсов (товаров/items) 

# Вопросы
#### Надо ли показывать Explore (surprise/diversity) рекомендации?
- например (только программистам) Dev курсы или курсы по Алгоритмам?
- надо ли (DS) специалистам показывать курсы по "сложной математике", например [Алгебраическая теория графов](https://openedu.ru/course/nsu/GRAPHS/)

#### Можно ли (Этично, Экономически оправдано):
- Для не прошедших курсы, но "отзывчивых" студентов рекомендовать предварительные курсы по отношению к тому на который они записались, но не прошли?
- Для "Холодных" показывать "популярные для входа (вкатиться)" в IT/DS курсы?
- Рекомендовать курсы по наиболее востребованным специальностям, например, golang специалистам. 

#### Кому какую скидку предлагать:
- тем кто заинтересован обучаться (есть весомые курсы в связках, которые он не прошел)
- у кого проблемы с оплатой полной стоимости были раньше,
- ...

#### Какие дополнительные бонусы от обучения:
- сообщество,
- консультации трудоустройства,
- ...
